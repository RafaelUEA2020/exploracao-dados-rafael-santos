```python
import pandas as pd

file_path = "/content/datatran_reduzed.csv"
df = pd.read_csv(file_path)

pd.options.display.float_format = '{:.2f}'.format

updatedf = df.describe()
updatedf.loc['var'] = df.var(numeric_only=True)

iqr = updatedf.loc['75%'] - updatedf.loc['25%']
updatedf.loc['iqr'] = iqr

updatedf
```

###O comando df.describre()

É um comando bem autoexplicativo, ele realiza a descrição da tabela trazendo os valores de Media, mediana, moda, minimo, maximo e algumas informações extras, Apenas dos valores que permitem tão cálculo, campos de caracteres não entram no calculo.
Foi utilizado um **format** para alterar a quantidade de casas decimais dos valores, pois o **decribe()** não faz o controle dessa quantidade.

###Valores

**count** conta a quantidade de valores.<br>
**mean** é o calculo da média dos valores.<br>
**std** é o desvio padrão dos valores.<br>
**min** Mostra o valor mínimo da coluna.<br>
**max** mostra o valor máximo da coluna.<br>
**var** é onde é mostrado a variância do valores da coluna, utilizando os valores da variável **std**.<br>
Os **intervalos quartis** na tabela são representados pelos valores **25%, 50%, 75%** cada um responsável por exibir qual é o valor que limita os outros valores menores aos quais estão sendo exibidos nas porcentagens.<br>
**iqr** é o valor do calculo do intervalor quartil de cada coluna.
